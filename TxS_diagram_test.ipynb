```python
#more complex - https://github.com/hafez-ahmad/TS_DIAGRAM_PY_R/blob/master/tsplot.py

ts=data_prec[["temperature", "Salinity", "precipitation"]]
ts.head()
```

```python
df=ts.sort_values("temperature",ascending=True)
df.head()
```

```python
#rang max and min, axis limits
mint=np.min(df["temperature"])
maxt=np.max(df["temperature"])

mins=np.min(df["Salinity"])
maxs=np.max(df["Salinity"])
```

```python
tempL=np.linspace(mint-1,maxt+1,786)
salL=np.linspace(mins-1,maxs+1,786)
```

```python
import gsw #density calculation
Tg, Sg = np.meshgrid(tempL,salL)
sigma_theta = gsw.sigma0(Sg, Tg)
cnt = np.linspace(sigma_theta.min(), sigma_theta.max(),786)
```

```python
fig,ax=plt.subplots(figsize=(10,10))
cs = ax.contour(Sg, Tg, sigma_theta, colors="grey", zorder=1)
cl=plt.clabel(cs,fontsize=10,inline=False,fmt="%.1f")

sc=plt.scatter(df["Salinity"],df["temperature"],c=df["precipitation"],s=10)

cb=plt.colorbar(sc)

ax.set_xlabel("Salinity")
ax.set_ylabel("Temperature[degC]")
ax.set_title("General T x S Diagram - HELGOLAND",fontsize=14, fontweight="bold")

ax.xaxis.set_major_locator(MaxNLocator(nbins=6))
ax.yaxis.set_major_locator(MaxNLocator(nbins=8))

ax.tick_params(direction="out")
cb.ax.tick_params(direction="out")
cb.set_label("Precipitation (mm)")
plt.tight_layout()
plt.show()
```
